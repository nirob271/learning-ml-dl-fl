# Python Programming Assignments-Conditional


1. Write a Python program that asks the user to input their age and prints whether they are eligible to vote (18 years or older) or not.

```python
n = int(input("Enter your age: "))
if n >= 18:
  print("User can vote")
else:
  print("User can not vote")
```

2. Write a Python program that checks if a user’s input number is positive, negative, or zero.

```python
n = int(input("Enter number :  "))
if n >= 0:
  if n == 0:
    print("Zero")
  else:
    print("Positive")
else:
  print("Negative")
```

3. write a program to determine if a number entered by the user is even or odd. The program should print 'Even Number' or 'Odd Number' based on the input.

```python
n = int(input("Enter number :  "))

if n%2==0:
  print("Even")
else:
  print("Odd")
```

4. Write a Python program that takes the user’s age and prints which age group they belong to:
- 'Child' for age 0–12
- 'Teenager' for age 13–19
- 'Adult' for age 20–64
- 'Senior' for age 65 and above

```python
n = int(input("Enter your age: "))

if n <= 12:
  print("Child")
elif n <= 19:
  print("Teenager")
elif n <= 64:
  print("Adult")
else:
  print("Senior")
```

# Python Programming Assignments - Loops and Functions

1. Write a Python function `sum_of_natural_numbers(n)` that returns the sum of all natural numbers up to `n`. Use a loop inside the function.

```python
n = int(input("Enter number: "))

sum = 0
for i in range(n+1):
  sum += i

print(sum);
```

Write a Python function `factorial(n)` that calculates and returns the factorial of `n`. Use a loop to calculate the factorial.

```python
n = int(input("Enter number: "))

F = 1

for i in range(1,n+1):
  F *= i

print(F)

```

3. Write a Python function `multiplication_table(n)` that prints the multiplication table of `n` up to 10.

```python
n = int(input("Enter number: "))

for i in range(1,11):
  print(f"{n} * {i} = {n*i}")
```
