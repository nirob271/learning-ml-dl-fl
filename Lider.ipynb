```python
%%writefile dataset.py
import os
import torch
import pandas as pd
import numpy as np
from torch.utils.data import Dataset

class TCGAPreExtractedDataset(Dataset):
    def __init__(self, omics_csv_path, wsi_dir_path):
        self.omics_data = pd.read_csv(omics_csv_path)
        self.wsi_dir_path = wsi_dir_path

        self.time_col = 'survival_months'
        self.event_col = 'censorship'

        numeric_cols = self.omics_data.select_dtypes(include=['number']).columns.tolist()

        self.feature_cols = [
            col for col in numeric_cols
            if col not in [self.time_col, self.event_col]
        ]

    def __len__(self):
        return len(self.omics_data)

    def __getitem__(self, idx):
        row = self.omics_data.iloc[idx]
        case_id = row['case_id']

        omics = row[self.feature_cols].values.astype(np.float32)
        omics_tensor = torch.tensor(omics)

        time_tensor = torch.tensor(np.float32(row[self.time_col]))
        event_tensor = torch.tensor(np.float32(row[self.event_col]))

        matches = [
            f for f in os.listdir(self.wsi_dir_path)
            if f.startswith(case_id) and f.endswith(".pt")
        ]

        if len(matches) > 0:
            pt_path = os.path.join(self.wsi_dir_path, matches[0])
            wsi_tensor = torch.load(pt_path)
            wsi_tensor = wsi_tensor.type(torch.float32)
        else:
            wsi_tensor = torch.zeros((1,1024), dtype=torch.float32)

        return omics_tensor, wsi_tensor, time_tensor, event_tensor


def collate_fn(batch):
    omics = torch.stack([item[0] for item in batch])
    wsis = [item[1] for item in batch]
    times = torch.stack([item[2] for item in batch])
    events = torch.stack([item[3] for item in batch])

    return omics, wsis, times, events
```

#Auto Create TransMIL Tuning Script

```python
%%writefile tune_transmil.py
import os
import torch
import torch.optim as optim
import pandas as pd
import numpy as np
from torch.utils.data import DataLoader, Subset

from dataset import TCGAPreExtractedDataset, collate_fn
from utils import cox_ph_loss, calculate_c_index

from models import TransMIL

def train_and_evaluate_fold(model, train_loader, test_loader, epochs=30, lr=1e-4, device='cuda'):
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-5)
    model = model.to(device)

    for epoch in range(epochs):
        model.train()
        for omics, wsis, times, events in train_loader:
            omics, times, events = omics.to(device), times.to(device), events.to(device)
            wsis = [w.to(device) for w in wsis]

            optimizer.zero_grad()
            risk_scores = model(omics, wsis)
            loss = cox_ph_loss(risk_scores, times, events)
            loss.backward()
            optimizer.step()

    model.eval()
    all_risks, all_times, all_events = [], [], []

    with torch.no_grad():
        for omics, wsis, times, events in test_loader:
            omics, times, events = omics.to(device), times.to(device), events.to(device)
            wsis = [w.to(device) for w in wsis]

            risk_scores = model(omics, wsis)

            all_risks.append(risk_scores)
            all_times.append(times)
            all_events.append(events)

    all_risks = torch.cat(all_risks)
    all_times = torch.cat(all_times)
    all_events = torch.cat(all_events)

    return calculate_c_index(all_risks, all_times, all_events)

def run_cross_validation(model_class, dataset, model_kwargs, splits_dir, epochs=30, device='cuda'):

    fold_results = []

    for fold in range(5):

        split_path = os.path.join(
            splits_dir,
            f"fold_{fold}",
            f"splits_{fold}.csv"
        )

        split_df = pd.read_csv(split_path)

        split_df.columns = split_df.columns.str.strip().str.lower()

        train_ids = split_df['train'].dropna().values
        test_ids = split_df['val'].dropna().values

        train_indices = dataset.omics_data.index[
            dataset.omics_data['case_id'].isin(train_ids)
        ].tolist()

        test_indices = dataset.omics_data.index[
            dataset.omics_data['case_id'].isin(test_ids)
        ].tolist()

        train_sub = Subset(dataset, train_indices)
        test_sub = Subset(dataset, test_indices)

        train_loader = DataLoader(
            train_sub,
            batch_size=32,
            shuffle=True,
            collate_fn=collate_fn
        )

        test_loader = DataLoader(
            test_sub,
            batch_size=32,
            shuffle=False,
            collate_fn=collate_fn
        )

        model = model_class(**model_kwargs)

        c_index = train_and_evaluate_fold(
            model,
            train_loader,
            test_loader,
            epochs=epochs,
            device=device
        )

        print(f"Fold {fold+1}: {c_index:.4f}")

        fold_results.append(c_index)

        if device == 'cuda':
            torch.cuda.empty_cache()

    print()
    print("Mean:", np.mean(fold_results))
    print("Std :", np.std(fold_results))

if __name__ == "__main__":

    DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'

    dataset = TCGAPreExtractedDataset(
        "./tcga_blca_all_clean.csv",
        "./tcga_blca_shrunk"
    )

    run_cross_validation(
        model_class=TransMIL,
        dataset=dataset,
        model_kwargs={'in_dim':1024},
        splits_dir="./splits/5foldcv/tcga_blca",
        epochs=30,
        device=DEVICE
    )
```
