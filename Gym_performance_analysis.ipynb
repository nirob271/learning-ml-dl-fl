#Dataset

```python
from google.colab import drive
drive.mount('/content/drive')
import warnings
warnings.filterwarnings('ignore')
gym = pd.read_csv('/content/drive/MyDrive/Dataset/gym_membership.csv')
```

#Understand The Data

```python
print(gym.head())
print(gym.shape)
print(gym.info())
print(gym.describe())
print(gym.columns.tolist())
```

#Bar Chart

```python
bar = gym['abonoment_type'].value_counts()
plt.bar(bar.index,bar.values,color=['red','green'])

plt.title('Gym Membership Type')
plt.xlabel('Abonoment Type')
plt.ylabel('Count')
plt.show()
```

#Histogram

```python
his = plt.hist(gym['avg_time_in_gym'])
plt.title('Average Time in Gym')
plt.xlabel('Average Time')
plt.ylabel('Count')
plt.show()
```

#Scatter plot

```python
sc = plt.scatter(gym['Age'],gym['visit_per_week'])
plt.title('Age vs Visit Per Week in Gym')
plt.xlabel('Age')
plt.ylabel('Visit Per Week')
plt.show()
```

# Box plot

```python
sns.boxplot(data=gym,x='gender',y='Age')
plt.title('Age Distribution by Gender')
plt.show()
```

# Count plot

```python
# sns.countplot(data=gym,x='days_per_week',order=sorted(gym['days_per_week'].unique()))
# plt.title('Count of Days per Week')
# plt.xticks(rotation=45)
# plt.show()

plt.figure(figsize=(35,8))
sns.countplot(data=gym, x='days_per_week', color='lightblue', edgecolor='black')
plt.title('Members vs Days Per Week')
plt.xlabel('Days Per Week')
plt.ylabel('No. of Members')
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.tight_layout()
plt.show()

```

#Bar plot

```python
sns.barplot(data=gym,x='abonoment_type', y='visit_per_week')
plt.title('Average Visit Per Week by Membership Type')
plt.xlabel('Membership Type')
plt.ylabel('Visit Per Week')
plt.show()
```
