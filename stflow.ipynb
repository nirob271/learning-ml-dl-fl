```python
def get_normalize_method(normalize_method, **kwargs):
    if normalize_method == "log1p":
        return log1p
    elif normalize_method == "stdiff":
        return stdiff_normalize
    elif normalize_method == "scVGAE":
        return scVGAE_normalize
```
