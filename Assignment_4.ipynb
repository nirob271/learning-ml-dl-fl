#Create a NumPy array with 20 random integers between 1 and 100 and Display the NumPy array

```python
import numpy as np

arr1 = np.array([1,3,2,4,6,5,8,7,0,9,11,15,32,25,87,65,36,25,31,9])
print(arr1)
```

# 1. Find the sum of all elements

```python
sum = np.sum(arr1)
print(sum)
```

# 2. Find the maximum and minimum values in the array

```python
mx = np.max(arr1)
mn = np.min(arr1)
print("Max Number: " , mx)
print("Min Number: " ,mn)
```

# Import Pandas

```python
import pandas as pd
```

# 1. Create a DataFrame with 5 rows and 3 columns of random numbers

```python
random_numbers = np.random.randint(1,101,size=(5,3))
df = pd.DataFrame(random_numbers, columns = ['A', 'B', 'C'])
print(df)
```

# 2. Compute the sum of each column


```python
col_sum = df.sum()

print("Sum of each column:")
print(col_sum)
```

# 3. Compute the mean of each column

```python
col_mn = df.mean()

print("Mean of each column:")
print(col_mn)
```
